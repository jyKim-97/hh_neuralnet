```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_dat(fname, N=2):
    with open(fname, "rb") as fp:
        data = np.fromfile(fp, dtype=np.double)
    return data.reshape([-1, N])
```

```python
vs = read_dat("./v_out.dat")
ics = read_dat("./i_app.dat")
syn_r = read_dat("./syn_r.dat")
syn_d = read_dat("./syn_d.dat")
ts = np.arange(vs.shape[0]) * 0.01
```

```python
plt.figure(dpi=120, figsize=(6, 2))
plt.plot(ts, ics[:,0], 'k', lw=1)
plt.plot(ts, ics[:,1], 'g', lw=1)
plt.show()
```

```python
nid = 1

plt.figure(dpi=120, figsize=(6,4), facecolor="w")
p1, = plt.plot(ts, syn_r[:,nid] - syn_d[:,nid], c='r', lw=2)
plt.ylabel(r"$\bar{g}$", fontsize=15, c='r')
plt.xlabel("Time (ms)", fontsize=15)

plt.twinx()
plt.plot(ts, vs[:,0], lw=2, c='k', label="pre")
plt.plot(ts, vs[:,1], lw=2, c='g', label="post")
plt.ylabel(r"$v$ (mV)", fontsize=15, c='k')
plt.legend(fontsize=12, edgecolor="none", loc="upper left")

plt.xlim([30, 100])
plt.tight_layout()
plt.savefig("./twosample_figure.png")
plt.show()
```
